```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv(r"X:\georisk\HaRIA_B_Wind\projects\acs\2. DATA\1. Work Unit Assessment\NWRA\hazard\aep_windspeed.csv")
```

```python
plt.semilogy(df[['A0-5', 'B1-2', 'C', 'D']], df['Annual Exceedance Probability'], label=['A0-5', 'B1-2', 'C', 'D'])
plt.ylabel("AEP")
plt.xlabel("Wind speed [m/s]")
plt.grid(linestyle='--')
plt.grid(which='minor', linestyle='--', alpha=0.5)
plt.legend()
```

```python
plt.semilogx(df['Return period'], df[['A0-5', 'B1-2', 'C', 'D']]*3.6, label=['A0-5', 'B1-2', 'C', 'D'])
plt.xlabel("Return period (years)")
plt.ylabel("Wind speed (km/h)")
plt.grid(linestyle='--')
plt.grid(which='minor', linestyle='--', alpha=0.5)
plt.legend()
```
